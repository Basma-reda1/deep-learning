```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))

def tanh(x):
    return math.tanh(x)

x = [1,2,3]

Wf, Whf = 0.5, 0.1
Wi, Whi = 0.6, 0.2
Wc, Whc = 0.7, 0.3
Wo, Who = 0.8, 0.4

bf = bi = bc = bo = 0

h = 0
c = 0

for t in range(len(x)):

    xt = x[t]
    print("\n==========")
    print("Time step:",t+1," Input:",xt)

    f = sigmoid(Wf*xt + Whf*h + bf)
    print("Forget gate =",round(f,4))

    i = sigmoid(Wi*xt + Whi*h + bi)
    print("Input gate =",round(i,4))

    c_tilde = tanh(Wc*xt + Whc*h + bc)
    print("Candidate =",round(c_tilde,4))

    c = f*c + i*c_tilde
    print("Cell state =",round(c,4))

    o = sigmoid(Wo*xt + Who*h + bo)
    print("Output gate =",round(o,4))

    h = o * tanh(c)
    print("Hidden state =",round(h,4))

Wy = 4
by = 0

y_hat = Wy * h + by

print("\n==========")
print("Final Prediction =",round(y_hat,2))
```
