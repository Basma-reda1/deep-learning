```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))

def tanh(x):
    return math.tanh(x)

# sequence with extra step
x = [1,2,3,4]

# weights from slide
Wf, Whf = 0.5, 0.1
Wi, Whi = 0.6, 0.2
Wc, Whc = 0.7, 0.3
Wo, Who = 0.8, 0.4

# biases
bf = bi = bc = bo = 0

# initial states
h = 0
c = 0

for t in range(len(x)):

    xt = x[t]

    f = sigmoid(Wf*xt + Whf*h)
    i = sigmoid(Wi*xt + Whi*h)
    c_tilde = tanh(Wc*xt + Whc*h)

    c = f*c + i*c_tilde

    o = sigmoid(Wo*xt + Who*h)
    h = o * tanh(c)

    print("t =",t+1)
    print("h =",round(h,3))
    print("c =",round(c,3))
    print()

# prediction layer
Wy = 4
b = 0

y_hat = Wy * h + b

print("Final prediction =",round(y_hat,3))
```
